```sql
SELECT DISTINCT
    c.cluster_id,
    c.cluster_name,
    c.delete_time,
    c.driver_node_type,
    c.worker_node_type,
    c.worker_count,
    c.auto_termination_minutes,
    c.cluster_source,
    MIN(n.start_time) AS first_start,
    MAX(n.end_time) AS last_start,
    w.workspace_name
FROM system.compute.clusters AS c
LEFT JOIN system.compute.node_timeline AS n
    ON c.cluster_id = n.cluster_id
LEFT JOIN system.access.workspaces_latest AS w
    ON w.workspace_id = n.workspace_id
WHERE driver = TRUE
  AND (:filterDeleted = 'No' OR c.delete_time IS NULL)
GROUP BY
    c.cluster_id,
    c.cluster_name,
    c.delete_time,
    c.driver_node_type,
    c.worker_node_type,
    c.worker_count,
    c.auto_termination_minutes,
    c.cluster_source,
    w.workspace_name
ORDER BY
    cluster_name
```
